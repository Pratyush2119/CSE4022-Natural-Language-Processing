```python
import nltk
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('averaged_perceptron_tagger')
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
```

**Loading the corpora -**

```python
with open('mycorpora.txt') as f:
  corpora = f.read()
```

```python
print(corpora)
```

**Word Segmentation of the corpus -**

```python
list_word=nltk.word_tokenize(corpora)
list_word
```

**Sentence segmentation of the corpus -**

```python
list_sent=nltk.sent_tokenize(corpora)
list_sent
```

**Converting to the lower case -**

```python
lowercase_words=[]
for sent in list_sent:
  for word in sent.split():
    lowercase_words.append(word.lower())

lowercase_words
```

**The number of words in the corpus -**

```python
len(lowercase_words)
```

**Removal of the Stop-words from the corpus -**

```python
stop_words = set(stopwords.words('english'))
filtered_words=[]
for word in lowercase_words:
  if word not in stop_words:
    filtered_words.append(word)

filtered_words
```

**Count of words after Stop-words removal -**

```python
len(filtered_words)
```

**Stemming of the filtered list of words -**

```python
ps = PorterStemmer()
for w in filtered_words:
  print(w," : ", ps.stem(w))
```

**Lemmatization of the filtered words -**

```python
lemmatizer = WordNetLemmatizer()
for w in filtered_words:
  print(w," : ", lemmatizer.lemmatize(w))
```

**Part of speech tagging of the filtered-words -**

```python
nltk.pos_tag(filtered_words)
```

**Part of Speech tagging for the list of all the words -**

```python
nltk.pos_tag(list_word)
```
